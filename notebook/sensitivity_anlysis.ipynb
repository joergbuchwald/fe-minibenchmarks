```python
def symb(model):
    if "trm" in model:
        return "b-"
    else:
        return "r--"
```

```python
for run in df_ts["GBM3"]["press"]:
    if "time" in run:
        pass
    else:
        plt.plot(df_ts["GBM1"]["temp"]["time"], df_ts["SC1"]["press"][run], symb(run))
```
